```python
#市ごとの医療福祉の売上金額
df_medical = pd.read_csv('FEI_MEDICAL.csv', encoding='SJIS', sep=',', quotechar='"', skiprows=8)
display(df_medical.head())
```

```python
#市ごとの総人口
df_people = pd.read_csv('FEI_PEOPLE.csv', encoding='SJIS', sep=',', quotechar='"', skiprows=8)
display(df_people.head())
```

```python
#市ごとの平均余命
df_death = pd.read_csv('FEI_DEATH.csv', encoding='SJIS', sep=',', quotechar='"', skiprows=8)
display(df_death.head())
```

```python
# '売上金額（医療、福祉）' 列を数値に変換します。変換できない値はNaNになります。
df_medical['C610120_売上金額（民営）（医療、福祉）【百万円】'] = pd.to_numeric(df_medical['C610120_売上金額（民営）（医療、福祉）【百万円】'], errors='coerce')

# 人口データも同様に数値に変換しておきましょう。
df_people['A1101_総人口【人】'] = pd.to_numeric(df_people['A1101_総人口【人】'], errors='coerce')

# 平均余命データも同様に数値に変換します。
df_death['I1101_平均余命（0歳）（男）【年】'] = pd.to_numeric(df_death['I1101_平均余命（0歳）（男）【年】'], errors='coerce')
df_death['I1102_平均余命（0歳）（女）【年】'] = pd.to_numeric(df_death['I1102_平均余命（0歳）（女）【年】'], errors='coerce')

# 変換後のデータ型を確認
print("医療福祉データ:")
display(df_medical.head())
print("\n人口データ:")
display(df_people.head())
print("\n平均余命データ:")
display(df_death.head())
```

```python
# 「売上金額（民営）（医療、福祉）【百万円】」列の欠損値の数を確認
missing_medical_sales_count = df_medical['C610120_売上金額（民営）（医療、福祉）【百万円】'].isnull().sum()
print(f"「売上金額（民営）（医療、福祉）【百万円】」列の欠損値の数: {missing_medical_sales_count}")

# 「平均余命（0歳）（男）【年】」列の欠損値の数を確認
missing_male_death_count = df_death['I1101_平均余命（0歳）（男）【年】'].isnull().sum()
print(f"「平均余命（0歳）（男）【年】」列の欠損値の数: {missing_male_death_count}")

# 「平均余命（0歳）（女）【年】」列の欠損値の数を確認
missing_female_death_count = df_death['I1102_平均余命（0歳）（女）【年】'].isnull().sum()
print(f"「平均余命（0歳）（女）【年】」列の欠損値の数: {missing_female_death_count}")
```

```python
# 不要な列を削除
df_medical = df_medical.drop(columns=['/項目', '注釈'])
df_people = df_people.drop(columns=['/項目', '注釈'])
df_death = df_death.drop(columns=['/項目', '注釈', '注釈.1'])

# 削除後のデータフレームの先頭を表示して確認
display(df_medical.head())
display(df_people.head())
display(df_death.head())
```

```python
# df_medical と df_people を、共通する列をキーとして内部結合
merged_df = pd.merge(df_medical, df_people, on=['調査年 コード', '調査年', '地域 コード', '地域'], how='inner')

# 結合後のデータフレームの最初の数行を表示して確認
display(merged_df.head())
```

## 一人当たり売上計算

```python
# 一人当たり医療福祉売上金額を計算
# 人口0での割り算を避けるため、割り算の前に人口0をNaNに置き換え、割り算後にNaNを0で埋める
merged_df['一人当たり医療福祉売上金額'] = (merged_df['C610120_売上金額（民営）（医療、福祉）【百万円】'] / merged_df['A1101_総人口【人】']).fillna(0)

# 新しい列が追加されたデータフレームの最初の数行を表示して確認
display(merged_df.head())
```

```python
merged_df.info()
display(merged_df.describe())
```

```python
# 新しい列 '平均余命（男女平均）【年】' として追加
df_death['平均余命（男女平均）【年】'] = (df_death['I1101_平均余命（0歳）（男）【年】'] + df_death['I1102_平均余命（0歳）（女）【年】']) / 2

display(df_death.head())
```

```python
# merged_df と df_death を、共通する列をキーとして内部結合
final_merged_df = pd.merge(merged_df, df_death[['調査年 コード', '地域 コード', '平均余命（男女平均）【年】']], on=['調査年 コード', '地域 コード'], how='inner')

# 結合後のデータフレームの最初の数行を表示して確認
display(final_merged_df.head())
```

```python
# 平均余命の男女平均ランキングを作成
life_expectancy_ranking = final_merged_df.sort_values(by='平均余命（男女平均）【年】', ascending=False)
print("\n平均余命（男女平均）ランキング:")
display(life_expectancy_ranking[['地域', '平均余命（男女平均）【年】']].head(10))
```

```python
# 市場規模（医療、福祉の売上金額）ランキングを作成
market_size_ranking = final_merged_df.sort_values(by='C610120_売上金額（民営）（医療、福祉）【百万円】', ascending=False)
print("\n市場規模（医療、福祉の売上金額）ランキング:")
display(market_size_ranking[['地域', 'C610120_売上金額（民営）（医療、福祉）【百万円】']].head(10))
```

```python
# 一人当たり医療福祉売上金額ランキングを作成
per_capita_ranking = final_merged_df.sort_values(by='一人当たり医療福祉売上金額', ascending=False)
print("\n一人当たり医療福祉売上金額ランキング:")
display(per_capita_ranking[['地域', '一人当たり医療福祉売上金額']].head(10))
```

```python
from scipy.stats import pearsonr
import pandas as pd

# 山形県寒河江市を含めた全データでの計算
# final_merged_df は既に結合済みのデータフレームです

# 欠損値を含む行を除外して計算
corr_all, p_value_all = pearsonr(final_merged_df['一人当たり医療福祉売上金額'].dropna(), final_merged_df['平均余命（男女平均）【年】'].dropna())
r_squared_all = corr_all**2

print("--- 山形県寒河江市を含めた全データ ---")
print(f"ピアソン相関係数 (r): {corr_all:.2f}")
print(f"決定係数 (R²): {r_squared_all:.2f}")
print(f"p値: {p_value_all:.3f}")

print("\n" + "="*30 + "\n")

# 山形県寒河江市を外れ値として除外したデータでの計算
# filtered_df は既に作成されています

# 欠損値を含む行を除外して計算
corr_filtered, p_value_filtered = pearsonr(filtered_df['一人当たり医療福祉売上金額'].dropna(), filtered_df['平均余命（男女平均）【年】'].dropna())
r_squared_filtered = corr_filtered**2

print("--- 山形県寒河江市を除外したデータ ---")
print(f"ピアソン相関係数 (r): {corr_filtered:.2f}")
print(f"決定係数 (R²): {r_squared_filtered:.2f}")
print(f"p値: {p_value_filtered:.3f}")
```
